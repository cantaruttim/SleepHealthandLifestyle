```python
f, ax = plt.subplots(figsize=(10, 4))

sns.despine(f, left=True, bottom=True)

sns.scatterplot(x="Quality of Sleep", y="Sleep Duration",
                sizes=(1, 8), linewidth=0,
                data=dataset, ax=ax);
```
